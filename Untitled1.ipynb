```python
def generate_ngrams(words_list, n):
    ngrams_list = []
 
    for num in range(0, len(words_list)):
        ngram = ' '.join( str(words_list[num:num + n]))
        ngrams_list.append(ngram)
 
    return ngrams_list
```

```python
import numpy as np
from collections import Counter
from itertools import chain

# Define the n-grams that we want to calculate mutual information for
n_grams = [1, 2]


# Extract the sequences and classes from the dataframe
sequences = df['Seq_Syllables']
classes = df['labels.label_context'].apply(lambda x : x[0])

# Step 2: Calculate the frequency of each n-gram in the dataset
n_gram_counts = {}
for n in n_grams:
    n_gram_counts.update(df['Seq_Syllables'].apply(lambda x :generate_ngrams(x, n )) )


# Step 3: Calculate the joint probability of each n-gram and class
joint_counts = {}
for c in set(classes):
    class_indices = np.where(classes == c)[0]
    class_sequences = [sequences[i] for i in class_indices]
    class_n_gram_counts = Counter(chain.from_iterable(class_sequences))
    for n in n_grams[1:]:
        class_n_gram_counts.update(zip(*(class_sequences[i:] for i in range(n))))
    for n_gram in n_gram_counts:
        joint_counts[(n_gram, c)] = class_n_gram_counts[n_gram]

# Step 4: Calculate the marginal probability of each n-gram and class
n_gram_probs = {n_gram: count / sum(n_gram_counts.values()) for n_gram, count in n_gram_counts.items()}
class_probs = {c: np.mean(classes == c) for c in set(classes)}

# Step 5: Calculate the mutual information between each n-gram and class
mi_scores = {}
for n_gram, c in joint_counts.keys():
    joint_prob = joint_counts[(n_gram, c)] / len(classes)
    n_gram_prob = n_gram_probs[n_gram]
    class_prob = class_probs[c]
    if joint_prob == 0:
        mi_scores[(n_gram, c)] = 0
    else:
        mi_scores[(n_gram, c)] = np.log2(joint_prob / (n_gram_prob * class_prob))

# Print the mutual information scores for each n-gram and class
for n in n_grams:
    print(f"Mutual Information for {n}-gram:")
    for c in set(classes):
        print(f"Class {c}:")
        for n_gram in n_gram_counts:
            if len(n_gram) == n:
                mi_score = mi_scores[(n_gram, c)]
                print(f"{n_gram}: {mi_score}")
```

```python
def gen_seq_markov(alphabet, probs, seq_len):
    """ Generate a set of sample sequences from a Markov model
    """
    sequence = list(
        np.random.choice(alphabet, p=np.sum(probs, axis=0) / np.sum(probs), size=1)
    )
    for i in tqdm(range(seq_len-1)):
        sequence.append(np.random.choice(alphabet, p=probs[:, sequence[-1]], size=1)[0])
    return sequence
```

## Autocorrelations

```python
from tqdm.auto import tqdm



def autocorrelate(y, range_ = range_):
    return [np.corrcoef(y[i:], y[:-i])[0][1] for i in tqdm(range_)]
```

```python
range_  = np.arange(1, 10, 1)
```

```python
sequence = gen_seq_markov(alphabet, probs, seq_len = 100000)
```

```python

# compute autocorrelation over the range of lags
autocorr = autocorrelate(sequence, range_)
```

```python
autocorr;
```

```python


# plot the autocorrelation in log-log and linear
fig, axs = plt.subplots(ncols=2, figsize=(10,5))
ax = axs[0]
ax.plot(range_, autocorr)
ax.set_xlabel('Sequential distance / lag')
ax.set_ylabel('Correlation')
ax.set_yscale('log')
ax.set_xscale('log')
ax.grid(which='both')

ax = axs[1]
ax.plot(range_, autocorr)
ax.set_xlabel('Sequential distance / lag')
ax.set_ylabel('Correlation')
ax.grid(which='both')
```

```python
# plot the absolute value of the autocorrelation in log-log and linear
fig, axs = plt.subplots(ncols=2, figsize=(10,5))
ax = axs[0]
ax.plot(range_, np.abs(autocorr))
ax.set_xlabel('Sequential distance / lag')
ax.set_ylabel('abs Correlation')
ax.set_yscale('log')
ax.set_xscale('log')
ax.grid(which='both')

ax = axs[1]
ax.plot(range_, np.abs(autocorr))
ax.set_xlabel('Sequential distance / lag')
ax.set_ylabel('abs Correlation')
ax.grid(which='both')
```
